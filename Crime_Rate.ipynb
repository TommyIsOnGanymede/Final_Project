## Filtering Data

In the context of analyzing foreclosed homes for sale across different neighborhoods, we've previously identified a list of neighborhoods of interest in our earlier analyses. This list encapsulates neighborhoods that are relevant to our current focus - those with foreclosed homes available for sale.

The purpose of this list is to filter our current Crime DataFrame. By applying this filter, we aim to refine the data presented in the subsequent bar graph, ensuring that it only encompasses information pertinent to our targeted neighborhoods. This step is crucial for maintaining the relevance and clarity of our data visualization, enabling us to draw meaningful insights regarding foreclosed homes across the identified neighborhoods.


```python
filtered_df = df[df.isin(neighborhoods_to_include)]
```

```python
# Creating a new figure with the desired size before plotting
plt.figure(figsize=(30, 10))

# Now plotting the number of delinquencies by neighborhood
filtered_df['neighborhood'].value_counts().plot(kind='bar', color='skyblue')

# Setting the labels and title
plt.xlabel('Neighborhood')
plt.ylabel('Number of crime reports')
plt.title('Number of Crime Reports by Neighborhood (2016 - 2022)')

# Adjusting the rotation and alignment of x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.show()
```

```python
neighborhoods_ofInterest = ["Central Business District", "Shadyside", "Central Oakland", "North Oakland", "West Oakland", "Squirrel Hill South","Squirrel Hill North","Point Breeze","Point Breeze North"]
filtered_neighborhoods = filtered_df[filtered_df['neighborhood'].isin(neighborhoods_ofInterest)]
neighborhood_counts = filtered_neighborhoods['neighborhood'].value_counts()

# Manually sum the counts for the Oakland neighborhoods
oakland_count = (neighborhood_counts.get('Central Oakland', 0) +
                 neighborhood_counts.get('North Oakland', 0) +
                 neighborhood_counts.get('West Oakland', 0))

print(f"Total count for Oakland: {oakland_count}")

# Manually sum the counts for the Squirrel Hill neighborhoods
squirrelHill_count = (neighborhood_counts.get('Squirrel Hill South', 0) +
                     neighborhood_counts.get('Squirrel Hill North', 0))

# Manually sum the counts for the Squirrel Hill neighborhoods
pointbreeze_count = (neighborhood_counts.get('Point Breeze', 0) +
                     neighborhood_counts.get('Point Breeze North', 0))



print(f"Total count for Point Breeze: {pointbreeze_count}")

# Print counts for Central Business District and Shadyside
cbd_count = neighborhood_counts.get('Central Business District', 0)
shadyside_count = neighborhood_counts.get('Shadyside', 0)


print(f"Total count for Central Business District: {cbd_count}")
print(f"Total count for Shadyside: {shadyside_count}")
```
