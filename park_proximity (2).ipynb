Lauren's metric: Number of Parks in Each Neighborhood

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('parks_proximity.csv') #read in dataset
park_count = df['neighborhood'].value_counts() #counts number of parks in each neighborhood

plt.figure(figsize=(18, 7))  # Sets the size of the graph
park_count.plot(kind='bar', color='skyblue') #creates a bar plot and sets color
plt.xlabel('Neighborhood') #labels x axis
plt.ylabel('Number of Parks') #labels y yaxis
plt.title('Number of Parks in Each Neighborhood in Pittsburgh') #gives the graph a title
plt.xticks(rotation=55, ha='right', fontsize = 8)  #sets rotation and font size of x axis labels for readability
plt.show() #prints the graph
```

This chart shows the number of parks in each neighborhood of Pittsburgh. The higher the number of parks, the better the neighborhood according to our standard. East Liberty has the most parks, but I realized that the dataset included traffic medians as a park, which would not impact housing prices, so I made a second chart that filtered these out.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('parks_proximity.csv') #read in dataset
park_df = df[df['type'] == 'Park'] #filters data to just be parks
park_count = park_df['neighborhood'].value_counts() #counts how many parks in each neighborhood


plt.figure(figsize=(18, 7))  # Sets the size of the graph
park_count.plot(kind='bar', color='skyblue') #creates a bar plot and sets color
plt.xlabel('Neighborhood') #labels x axis
plt.ylabel('Number of Parks') #labels y yaxis
plt.title('Number of Parks in Each Neighborhood in Pittsburgh') #gives the graph a title
plt.xticks(rotation=55, ha='right', fontsize = 8)  #sets rotation and font size of x axis labels for readability
plt.show() #prints the graph
```

This chart sorts the data so that only true parks are accounted for. According to the new chart, Beechview and South Side Slopes have the most parks. This means that houses in these neighborhoods will generally have a higher quality of life due to the proximity to more parks.
