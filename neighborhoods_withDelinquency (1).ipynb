## Neighborhoods Where Households are Delinquent on their Property Taxes
    # Sorting Data
In order to make a bar-chart that best illustrates the data we are trying to show. We must first filter out the rows of data that show a household that owes $0 and/or has owed money for less than a year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading in the dataset
df = pd.read_csv('Tax_DelinquencyData.csv')
# Assigning variable names to columns
df.columns.values[7] = 'neighborhood'

# Filtering the DataFrame
filtered_df = df[(df['current_delq'] > 0) & (df['prior_years'] >= 1)]

# Creating a new figure with the desired size before plotting
plt.figure(figsize=(30, 10))

# Now plotting the number of delinquencies by neighborhood
filtered_df['neighborhood'].value_counts().plot(kind='bar', color='skyblue')

# Setting the labels and title
plt.xlabel('Neighborhood')
plt.ylabel('Count of Delinquencies')
plt.title('Number of Tax Delinquencies by Neighborhood')

# Adjusting the rotation and alignment of x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.show()
```

## Highligting The Top + Bottom 5 Data Points

```python
# Get the top 5 neighborhoods based on frequency
top_five_neighborhoods = filtered_df['neighborhood'].value_counts().head(5).index.tolist()
top_five_neighborhoods
```

```python
# Get the top 5 neighborhoods based on frequency
bottom_five_neighborhoods = filtered_df['neighborhood'].value_counts().tail(5).index.tolist()
bottom_five_neighborhoods
```
